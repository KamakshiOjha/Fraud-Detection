```python
import numpy as np
import pandas as pd #for CSV
from sklearn.model_selection import train_test_split #allows to split training and testing data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load to dataframe(pandas)
credit_data = pd.read_csv("/Users/kamakshiojha/Documents/Machine_Learning_projects/Fraud_Detection/creditcard.csv")
```

```python
#check missing values in each colm

credit_data.isnull().sum()
```

```python
# distribution of legit trans & fraud trans

credit_data['Class'].value_counts()
```

```python
#this ratio above is very low for fraud transaction - prediction will always show that it is legit 
#this is unbalanced dataset

#seperating data

legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]

print(legit.shape)
print(fraud.shape)
```

```python
#statistics for data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values for each class
credit_data.groupby('Class').mean()
```

```python
# dealing with unbalanced data - Under sampling

#build sample dataset for this dataset - for same distribution

legit_sample = legit.sample(n=492)
```

```python
#concat 2 data frames

new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#split data

X = new_dataset.drop(columns ='Class',axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#split into train and test

X_train,X_test ,Y_train ,Y_test = train_test_split(X,Y,test_size = 0.2, stratify =Y, random_state = 2)  
```

```python
print(X.shape , X_train.shape,X_test.shape)
```

```python
#train model - logistic resgression

model = LogisticRegression()

```

```python
#training the model

model.fit(X_train,Y_train)
```

```python
#evaluation of model

X_train_prec = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prec,Y_train)
```

```python
print(training_data_accuracy)
```

```python
# accuracy for test data

X_test_prec = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prec,Y_test)
```

```python
print(test_data_accuracy)
```
